<h2 style="border-left:15px solid orange;">Gradient Decent</h2>

Let’s implement gradient descent for L(w)=w<sup>2</sup>
</br>
<u>Update Rule:</u></br>
<img src = "image.png" alternate = "update rule">

```python
w = 5.0
lr = 0.1

for i in range(50):
    grad = 2 * w   # derivative
    w = w - lr * grad
    
    print(f"Step {i+1}: w = {w:.7f}")
```

<h4 style="border-left:10px solid green;">Batch Gradient Decent</h4>

<li>Batch Gradient Descent computes the gradient using the entire dataset at once before updating weights.</li><br>
<u>Update Rule</u><br>
<img src = "image copy.png" alternate="Update rule"><br>
<p>Let us implement Batch gradient decent using MSE as loss function and y_pred = wx as predictor</p>

```python
import numpy as np

# Dataset
X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 6, 8, 10], dtype=float)  # y = 2x

# Initialize weight
w = 0.0
lr = 0.01
epochs = 50

for epoch in range(epochs):
    # Predictions for ALL data
    y_pred = w * X
    
    # Compute gradient (MSE derivative)
    grad = (2/len(X)) * np.sum((y_pred - y) * X)
    
    # Update weight
    w = w - lr * grad
    
    print(f"Epoch {epoch+1}: w = {w:.4f}")
```

<h4 style="border-left:10px solid green;"> Stochastic Gradient Decent</h4>
<li>SGD updates the weights using only one training example at a time.</li>
<u>Upadate Rule:</u><br>
<img src = "image copy 3.png" alternative ="Upadte Rule">
<p>Let us implement Batch gradient decent using MSE as loss function and y_pred = wx as predictor</p>


```python
import numpy as np

# Dataset
X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 6, 8, 10], dtype=float)

# Initialize
w = 10
lr = 0.01
epochs = 10

for epoch in range(epochs):
    # Shuffle data
    indices = np.random.permutation(len(X)) # for shuffling the data in each epoch
    X_shuffled = X[indices] #fancy indexing retrival
    y_shuffled = y[indices]

    for i in range(len(X)):
        xi = X_shuffled[i]
        yi = y_shuffled[i]

        # Prediction
        y_pred = w * xi

        # Gradient (for ONE sample)
        grad = 2 * (y_pred - yi) * xi

        # Update immediately
        w = w - lr * grad
        print(f"{w:.4f}")

    print(f"Epoch {epoch+1}: w = {w:.4f}")
```

<li>This is called Stocastic due to it adds randomness to the optimizationa and somewhat helps finding solution non linearly which somewhat prevents local minima</li>
<li>Stochastic Gradient Descent updates weights using one sample at a time, making learning faster but noisier, and often helping the model escape poor solutions.</li>
<li>SGD trades precision for speed—and that randomness is often what makes deep learning work.</li>



<h4 style="border-left:10px solid green;"> Mini-Batch Gradient Descent </h4>
<li><p>Mini-batch gradient descent updates weights using small groups of data, giving a balance between stability (like batch GD) and speed (like SGD), while reducing the problem of opposing updates cancelling each other.</p></li>
<b>Update Rule</b><br>
<img src = "image copy 4.png">
<br>
<p>Now let us Implement the Mini Batch gradient decent with mse as loss function and y = 2x as learnable function</p>


```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output

# Dataset
X = np.array([1,2,3,4,5,6,7,8], dtype=float)
y = 2 * X

w = 100.0
lr = 0.01
batch_size = 2
epochs = 10

for epoch in range(epochs):
    indices = np.random.permutation(len(X))
    
    for i in range(0, len(X), batch_size):
        batch_idx = indices[i:i+batch_size]
        X_batch = X[batch_idx]
        y_batch = y[batch_idx]

        y_pred = w * X_batch

        grad = (2/len(X_batch)) * np.sum((y_pred - y_batch) * X_batch)

        w = w - lr * grad

    # y_pred = w*X

    # clear_output(wait=True)  # 🔥 important

    # plt.figure()
    # plt.scatter(X, y, alpha=0.2)
    # plt.plot(X, y_pred)

    # plt.title(f"Epoch {epoch+1}")
    # plt.show()
    #visualisation is not really does any helpp due to we are tryiing to learn the topic with simple examples.
    print(f"Epoch {epoch+1}: w = {w:.4f}")
```

<p>The speed of Batch Gradient and the randomness of stochastic gradient both are takes part in mini-batch optimization</p>

<p>Now let us Implement the Stochastic gradient decent with momentum with mse as loss function and y = 2x as learnable function with the both update rules which have the same effect</p>


```python
import numpy as np

# Dataset
X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 6, 8, 10], dtype=float)

# Initialize
w = 0.0
lr = 0.01
beta = 0.8 #0.9
v = 0.0
epochs = 20

for epoch in range(epochs):
    indices = np.random.permutation(len(X))
    
    for i in indices:
        xi = X[i]
        yi = y[i]

        y_pred = w * xi
        grad = 2 * (y_pred - yi) * xi

        # Momentum update
        v = beta * v + (1 - beta) * grad
        w = w - lr * v

    print(f"Epoch {epoch+1}: w = {w:.4f}")
```

```python
import numpy as np

# Dataset
X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 6, 8, 10], dtype=float)

# Initialize
w = 0.0
lr = 0.01
beta = 0.8 #0.9
v = 0.0
epochs = 20

for epoch in range(epochs):
    indices = np.random.permutation(len(X))
    
    for i in indices:
        xi = X[i]
        yi = y[i]

        y_pred = w * xi
        grad = 2 * (y_pred - yi) * xi

        # Momentum update
        v = beta * v - lr * grad
        w = w + v

    print(f"Epoch {epoch+1}: w = {w:.4f}")
```

<h4 style="border-left:10px solid green;">Nesterov's Accelerated Gradient</h4>
<li>Nesterov Accelerated Gradient improves momentum by computing the gradient at the predicted future position, allowing the model to correct its direction earlier and avoid overshooting.</li>
<img src = "image copy 8.png">

```python
import numpy as np

X = np.array([1,2,3,4,5], dtype=float)
y = 2 * X

w = 0.0
v = 0.0
lr = 0.01
beta = 0.9
epochs = 20

for epoch in range(epochs):
    indices = np.random.permutation(len(X)) 
    
    for i in indices:
        xi = X[i]
        yi = y[i]

        # Look-ahead
        w_look = w + beta * v

        # Gradient at look-ahead
        y_pred = w_look * xi
        grad = 2 * (y_pred - yi) * xi

        # Update
        v = beta * v - lr * grad
        w = w + v

    print(f"Epoch {epoch+1}: w = {w:.4f}")
```

<h2 style="border-left:15px solid orange;">Adaptive Learning Rate Methods</h2>

<h4 style="border-left:10px solid green;">AdaGrad (Adaptive Gradient)</h4>
<li>AdaGrad adapts the learning rate for each parameter by dividing the gradient by the accumulated squared gradients, making frequently updated parameters learn slower and infrequent ones learn faster.</li>
<img src="image copy 9.png"><br>
Parameter that changes a lot → slow it down

Parameter that rarely changes → allow bigger steps

```python
import numpy as np

X = np.array([1,2,3,4,5], dtype=float)
y = 2 * X

w = 0.0
lr = 0.1
G = 0.0
epsilon = 1e-8

for epoch in range(250):
    indices = np.random.permutation(len(X))
    
    for i in indices:
        xi = X[i]
        yi = y[i]

        y_pred = w * xi
        grad = 2 * (y_pred - yi) * xi

        # accumulate squared gradients
        G += grad**2

        # update
        w = w - (lr / (np.sqrt(G) + epsilon)) * grad

    print(f"Epoch {epoch+1}: w = {w:.4f}")
```

<li>G increases Forever and Learning rate decreasing forever</li>
<li>Learning Becomes slow and almost stops</li>

<h4 style="border-left:10px solid green;">RMSProp (Root Mean Square Propagation)</h4>
<li>RMSProp adapts the learning rate like AdaGrad but uses a moving average of recent squared gradients instead of accumulating all past gradients, preventing the learning rate from shrinking to zero.</li>
<img src="image copy 10.png">

```python
import numpy as np

X = np.array([1,2,3,4,5], dtype=float)
y = 2 * X

w = 0.0
lr = 0.01
S = 0.0
beta = 0.9
epsilon = 1e-8

for epoch in range(100):
    indices = np.random.permutation(len(X))
    
    for i in indices:
        xi = X[i]
        yi = y[i]

        y_pred = w * xi
        grad = 2 * (y_pred - yi) * xi

        # update moving average
        S = beta * S + (1 - beta) * (grad ** 2)

        # update weight
        w = w - (lr / (np.sqrt(S) + epsilon)) * grad

    print(f"Epoch {epoch+1}: w = {w:.4f}")
```

<h4 style="border-left:10px solid green;">AdaM (Adaptive Moment Estimation)</h4>
<li>Adam combines momentum (for stable direction) and RMSProp (for adaptive step size), making it a powerful and widely used optimizer for training neural networks efficiently.</li>
<img src ="image copy 11.png">

```python
import numpy as np

X = np.array([1,2,3,4,5], dtype=float)
y = 2 * X

w = 0.0
m = 0.0
v = 0.0

lr = 0.01
beta1 = 0.9
beta2 = 0.999
epsilon = 1e-8

t = 0

for epoch in range(200):
    indices = np.random.permutation(len(X))
    
    for i in indices:
        t += 1 
        #t is a global step counter used for bias correction, so resetting it every epoch would break the consistency between accumulated gradients and their correction.
        # for 20 epochs withs 50 updates each epoch the t values will be 1000 at the end
        
        xi = X[i]
        yi = y[i]

        y_pred = w * xi
        grad = 2 * (y_pred - yi) * xi

        # update moments
        m = beta1 * m + (1 - beta1) * grad
        v = beta2 * v + (1 - beta2) * (grad ** 2)

        # bias correction
        m_hat = m / (1 - beta1**t)
        v_hat = v / (1 - beta2**t)

        # update weight
        w = w - (lr / (np.sqrt(v_hat) + epsilon)) * m_hat

    print(f"Epoch {epoch+1}: w = {w:.4f}")

#Adam decides both where to go and how fast to go, using past gradients.    
```

<b>Cons of Adam</b>
<li>Can converge to sharp minima

<li>Generalization worse than SGD</li>

<li>Weight decay behaves incorrectly</li>

<h4 style="border-left:10px solid green;">AdaMW (Adaptive Moment Estimation with Weight decay)</h4>
<li>AdamW adds a little correction to the normal Adam , which is the Weight decay.</li>
<li>Without Weight decay the weights will become sharp and very sensitive to input and will learn noise.</li>
<li> Weight decay means add penalty to weights which gradually pulls the weights towards and reguralise them and deals with the un-generalizability problem of Adam</li>
<li>Noramally we add a +&#955||w||<sup>2</sup> term to the loss as weight decay  L<sub>new</sub> = L +&#955||w||<sup>2</sup></li>
<img src="image copy 13.png">
<br><br>
<li>But in AdamW we just change the update rule as follows</li>
<li>Just a (- &#951&#955w<sub>t</sub>) term to the Update rule and remaining all same as Adam</li> 
<img src="image copy 12.png">
<li><b>Implementation:</b></li>

```python
import numpy as np

# sample data
X = np.array([1,2,3,4,5], dtype=float)
y = 2 * X

# initialize
w = 0.0
m = 0.0
v = 0.0

lr = 0.01
beta1 = 0.9
beta2 = 0.999
epsilon = 1e-8
weight_decay = 0.01

t = 0

for epoch in range(2000):
    indices = np.random.permutation(len(X))
    
    for i in indices:
        t += 1
        
        xi = X[i]
        yi = y[i]

        # prediction
        y_pred = w * xi
        
        # gradient
        grad = 2 * (y_pred - yi) * xi
        
        # moments
        m = beta1 * m + (1 - beta1) * grad
        v = beta2 * v + (1 - beta2) * (grad ** 2)
        
        # bias correction
        m_hat = m / (1 - beta1**t)
        v_hat = v / (1 - beta2**t)
        
        # AdamW update
        w = w - (lr / (np.sqrt(v_hat) + epsilon)) * m_hat - lr * weight_decay * w

    print(f"Epoch {epoch+1}: w = {w:.4f}")
```

<li>We can see it staying Generalized even for 2000 epochs of training and this helps us to prevent overfitting</li>

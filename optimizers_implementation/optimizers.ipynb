<h3>Gradient Decent</h3>

Let’s implement gradient descent for L(w)=w<sup>2</sup>
</br>
<u>Update Rule:</u></br>
<img src = "image.png" alternate = "update rule">

```python
w = 5.0
lr = 0.1

for i in range(50):
    grad = 2 * w   # derivative
    w = w - lr * grad
    
    print(f"Step {i+1}: w = {w:.7f}")
```

<h4>Batch Gradient Decent</h4>

<li>Batch Gradient Descent computes the gradient using the entire dataset at once before updating weights.</li><br>
<u>Update Rule</u><br>
<img src = "image copy.png" alternate="Update rule"><br>
<p>Let us implement Batch gradient decent using MSE as loss function and y_pred = wx as predictor</p>

```python
import numpy as np

# Dataset
X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 6, 8, 10], dtype=float)  # y = 2x

# Initialize weight
w = 0.0
lr = 0.01
epochs = 50

for epoch in range(epochs):
    # Predictions for ALL data
    y_pred = w * X
    
    # Compute gradient (MSE derivative)
    grad = (2/len(X)) * np.sum((y_pred - y) * X)
    
    # Update weight
    w = w - lr * grad
    
    print(f"Epoch {epoch+1}: w = {w:.4f}")
```

<h4> Stochastic Gradient Decent</h4>
<li>SGD updates the weights using only one training example at a time.</li>
<u>Upadate Rule:</u><br>
<img src = "image copy 3.png" alternative ="Upadte Rule">
<p>Let us implement Batch gradient decent using MSE as loss function and y_pred = wx as predictor</p>


```python
import numpy as np

# Dataset
X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 6, 8, 10], dtype=float)

# Initialize
w = 0.0
lr = 0.01
epochs = 20

for epoch in range(epochs):
    # Shuffle data
    indices = np.random.permutation(len(X)) # for shuffling the data in each epoch
    X_shuffled = X[indices] #fancy indexing retrival
    y_shuffled = y[indices]

    for i in range(len(X)):
        xi = X_shuffled[i]
        yi = y_shuffled[i]

        # Prediction
        y_pred = w * xi

        # Gradient (for ONE sample)
        grad = 2 * (y_pred - yi) * xi

        # Update immediately
        w = w - lr * grad

    print(f"Epoch {epoch+1}: w = {w:.4f}")
```
